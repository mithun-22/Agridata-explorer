```python
# Clean column names: remove spaces, convert to lowercase, and replace spaces with underscores
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Print cleaned column names to verify
print(df.columns)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Compute correlation matrix
correlation_matrix = numeric_df.corr()

# Set figure size to improve readability
plt.figure(figsize=(20, 15))  # Increase figure size

# Draw heatmap with better settings
sns.heatmap(
    correlation_matrix,
    annot=False,   # Hide numbers to avoid clutter
    cmap='coolwarm',
    fmt=".2f",
    linewidths=0.5  # Add grid lines
)

# Improve x/y-axis readability
plt.xticks(rotation=90, fontsize=12)  # Rotate x-axis labels
plt.yticks(fontsize=12)               # Adjust y-axis font size
plt.title('Correlation Heatmap', fontsize=16)

plt.show()
```
