```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# 1. Top 7 RICE PRODUCTION State Data (Bar Plot)
top_rice = df.groupby("State Name")["RICE PRODUCTION (1000 tons)"].sum().sort_values(ascending=False).head(7)
sns.barplot(x=top_rice.values, y=top_rice.index, palette="YlGn")
plt.title("Top 7 Rice Producing States")
plt.xlabel("Rice Production (1000 tons)")
plt.ylabel("State")
plt.show()
```

```python
# 2. Top 5 Wheat Producing States (Bar Chart and Pie Chart)
top_wheat = df.groupby("State Name")["WHEAT PRODUCTION (1000 tons)"].sum().sort_values(ascending=False).head(5)

# Bar Chart
sns.barplot(x=top_wheat.values, y=top_wheat.index, palette="coolwarm")
plt.title("Top 5 Wheat Producing States")
plt.xlabel("Wheat Production (1000 tons)")
plt.show()

# Pie Chart
plt.pie(top_wheat.values, labels=top_wheat.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("coolwarm", 5))
plt.title("Wheat Production Share by Top 5 States")
plt.axis("equal")
plt.show()
```

```python
# 3. Oilseed production by top 5 states
top_oilseed = df.groupby("State Name")["OILSEEDS PRODUCTION (1000 tons)"].sum().sort_values(ascending=False).head(5)
sns.barplot(x=top_oilseed.values, y=top_oilseed.index, palette="viridis")
plt.title("Top 5 Oilseed Producing States")
plt.xlabel("Oilseed Production (1000 tons)")
plt.show()
```

```python
# 4. Top 7 SUNFLOWER PRODUCTION States
top_sunflower = df.groupby("State Name")["SUNFLOWER PRODUCTION (1000 tons)"].sum().sort_values(ascending=False).head(7)
sns.barplot(x=top_sunflower.values, y=top_sunflower.index, palette="autumn")
plt.title("Top 7 Sunflower Producing States")
plt.xlabel("Sunflower Production (1000 tons)")
plt.show()
```

```python
# 5. India's SUGARCANE PRODUCTION From Last 50 Years (Line Plot)
sugarcane_yearly = df.groupby("Year")["SUGARCANE PRODUCTION (1000 tons)"].sum()
sugarcane_yearly.plot(marker='o')
plt.title("India's Sugarcane Production (Last 50 Years)")
plt.xlabel("Year")
plt.ylabel("Production (1000 tons)")
plt.grid(True)
plt.show()
```

```python
# 6. Rice Production Vs Wheat Production (Last 50y)
prod_yearly = df.groupby("Year")[["RICE PRODUCTION (1000 tons)", "WHEAT PRODUCTION (1000 tons)"]].sum()
prod_yearly.plot()
plt.title("Rice vs Wheat Production Over Years")
plt.xlabel("Year")
plt.ylabel("Production (1000 tons)")
plt.grid(True)
plt.show()
```

```python
# 7. Rice Production By West Bengal Districts
wb_rice = df[df["State Name"] == "West Bengal"].groupby("Dist Name")["RICE PRODUCTION (1000 tons)"].sum().sort_values(ascending=False)
wb_rice.plot(kind='bar')
plt.title("Rice Production by Districts in West Bengal")
plt.ylabel("Production (1000 tons)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# 8. Top 10 Wheat Production Years From UP
up_wheat = df[df["State Name"] == "Uttar Pradesh"].groupby("Year")["WHEAT PRODUCTION (1000 tons)"].sum().sort_values(ascending=False).head(10)
sns.barplot(x=up_wheat.index, y=up_wheat.values, palette="Blues")
plt.title("Top 10 Years of Wheat Production in UP")
plt.ylabel("Wheat Production (1000 tons)")
plt.show()
```

```python
# 9. Millet Production (Last 50y) (Includes Pearl, Finger, and Total Millets)
df["MILLET PRODUCTION"] = df["PEARL MILLET PRODUCTION (1000 tons)"] + df["FINGER MILLET PRODUCTION (1000 tons)"]
millet_trend = df.groupby("Year")["MILLET PRODUCTION"].sum()
millet_trend.plot()
plt.title("Millet Production Over Years")
plt.xlabel("Year")
plt.ylabel("Production (1000 tons)")
plt.grid(True)
plt.show()
```

```python
# 10. Sorghum Production (Kharif and Rabi) by Region
sorghum = df.groupby("State Name")[["KHARIF SORGHUM PRODUCTION (1000 tons)", "RABI SORGHUM PRODUCTION (1000 tons)"]].sum()
sorghum.plot(kind='bar', stacked=True)
plt.title("Kharif vs Rabi Sorghum Production by State")
plt.ylabel("Production (1000 tons)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# 11. Top 7 States for Groundnut Production
top_groundnut = df.groupby("State Name")["GROUNDNUT PRODUCTION (1000 tons)"].sum().sort_values(ascending=False).head(7)
sns.barplot(x=top_groundnut.values, y=top_groundnut.index, palette="magma")
plt.title("Top 7 Groundnut Producing States")
plt.xlabel("Groundnut Production (1000 tons)")
plt.show()
```

```python
# 12. Soybean Production by Top 5 States and Yield Efficiency
top_soy = df.groupby("State Name")[["SOYABEAN PRODUCTION (1000 tons)", "SOYABEAN YIELD (Kg per ha)"]].sum()
top_soy_prod = top_soy.sort_values("SOYABEAN PRODUCTION (1000 tons)", ascending=False).head(5)
top_soy_prod[["SOYABEAN PRODUCTION (1000 tons)"]].plot(kind='bar', legend=False)
plt.title("Top 5 Soybean Producing States")
plt.ylabel("Production (1000 tons)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_soy_prod[["SOYABEAN YIELD (Kg per ha)"]].plot(kind='bar', color='orange', legend=False)
plt.title("Soybean Yield Efficiency of Top 5 States")
plt.ylabel("Yield (Kg per ha)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 13. Oilseed Production in Major States
major_oil = df.groupby("State Name")["OILSEEDS PRODUCTION (1000 tons)"].sum().sort_values(ascending=False).head(10)
major_oil.plot(kind='bar', color='olive')
plt.title("Oilseed Production in Major States")
plt.ylabel("Production (1000 tons)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 14. Impact of Area Cultivated on Production (Rice, Wheat, Maize)
plt.scatter(df["RICE AREA (1000 ha)"], df["RICE PRODUCTION (1000 tons)"], alpha=0.5, label='Rice')
plt.scatter(df["WHEAT AREA (1000 ha)"], df["WHEAT PRODUCTION (1000 tons)"], alpha=0.5, label='Wheat')
plt.scatter(df["MAIZE AREA (1000 ha)"], df["MAIZE PRODUCTION (1000 tons)"], alpha=0.5, label='Maize')
plt.legend()
plt.title("Area vs Production (Rice, Wheat, Maize)")
plt.xlabel("Area (1000 ha)")
plt.ylabel("Production (1000 tons)")
plt.grid(True)
plt.show()
```

```python
# 15. Rice vs. Wheat Yield Across States
yield_comp = df.groupby("State Name")[["RICE YIELD (Kg per ha)", "WHEAT YIELD (Kg per ha)"]].mean().dropna()
yield_comp.plot(kind='bar')
plt.title("Average Rice vs Wheat Yield Across States")
plt.ylabel("Yield (Kg per ha)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
